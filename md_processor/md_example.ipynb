# Market Data Example

### Quick Notebook example of using panadas and ploting

Load the data from the csv and plot mid price with buy/sell trades

```python
%matplotlib inline
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

pd.options.display.max_columns=40
pd.options.display.max_rows=1000
```

```python
path = 'data/md-test-2.C-M.csv'
book_data = pd.read_csv(path, index_col=False,header=None,nrows=50000)
```

```python
ask_trade_prices = book_data[book_data.iloc[:,0]=='S'].iloc[:,1]
bid_trade_prices = book_data[book_data.iloc[:,0]=='B'].iloc[:,1]
```

```python
book=book_data[book_data.iloc[:,0].isnull()]
book['mid_price']=(book.iloc[:,5]+book.values[:,6])/2.0
```

```python
plt.figure(figsize=(10,5))
book['mid_price'].plot(color='g')
plt.scatter(ask_trade_prices.index,ask_trade_prices.values,color='r',s=5)
plt.scatter(bid_trade_prices.index,bid_trade_prices.values,color='b',s=5)
```
